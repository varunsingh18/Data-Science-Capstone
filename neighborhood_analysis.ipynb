```python
#importing data

page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M' 
source = requests.get(page).text
soup = BeautifulSoup(source)

table_data = soup.find('div', class_='mw-parser-output')
table = table_data.table.tbody

#declaring columns
columns = ['PostalCode', 'Borough', 'Neighbourhood']

#create a dictonary
data = dict({key:[]*len(columns) for key in columns})

for row in table.find_all('tr'):
    for i,column in zip(row.find_all('td'),columns):
        i = i.text
        i = i.replace('\n', '')
        data[column].append(i)

#converting data into a pandas dataframe
df_pc = pd.DataFrame.from_dict(data=data)[columns]
```

```python
#cleaning dataframe
df_pc = df_pc[df_pc['Borough'] != 'Not assigned'].reset_index(drop = True)
print('After dropping rows where borough is "Not assigned", Shape is: ',df_pc.shape)
print('Number of rows where Neighbourhood is "Not assigned" but borough has value: ', 
      df_pc[df_pc['Neighbourhood'] == 'Not assigned'].shape[0])
```
